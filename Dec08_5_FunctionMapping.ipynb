# 함수 매핑 (mapping)

Series / DataFrame의 개별 요소를 특정 함수에 1:1로 대응시키는 과정

사용자가 직접 만든 함수 (람다 포함)를 적용시킬 수 있기 때문에,

for문이나 기본 함수로 처리하는 것 보다 훨씬 효율적

```python
# seaborn에서 타이타닉 데이터 불러와서 'age', 'fare'열만 인덱싱
#   => 'ten' 열을 만들어 10 넣기

t = sns.load_dataset('titanic')
df = t.loc[:, ['age', 'fare']]
df['ten'] = 10
df.head()
```

```python
# 값을 넣으면 10을 더해주는 함수
def plus_ten(a):
  return a + 10

# 값을 2개 넣으면 그 둘을 곱해주는 함수
def mult(a, b):
  return a * b

# 값을 3개 넣으면 그 셋을 더해주는 함수
def plus_triple(a, b, c):
  return a + b + c

print(plus_ten(5))
print(mult(1, 5))
print(plus_triple(1, 2, 3))
```

```python
# 함수들을 apply()를 이용해 시리즈 객체의 개별 요소에 적용
sr1 = df['age'].apply(plus_ten)
sr1.head()
```

```python
sr2 = df['fare'].apply(mult, b=0.9)    
# 첫번째 값은 앞의 fare열, 두번째 값은 지정
sr2.head()
```

```python
sr3 = df['fare'].apply(plus_triple, b=10.1 , c=20)
sr3.head()
```

```python
# 'ten'열의 값들을 lambda함수를 이용해서 add_10함수 적용시키기
sr4 = df['ten'].apply(lambda a : plus_ten(a))
sr4.head()
```

# apply() vs applymap()


```python
df_map = df.applymap(plus_ten) # 데이터 프레임의 모든 요소에 함수를 적용시킴
df_map
```
