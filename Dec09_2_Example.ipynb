```python
from bs4 import BeautifulSoup
import urllib.request as req
import requests
import pandas as pd
```

```python
# https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=cs205

# 전체 내용 데이터 parsing

# parser 종류 : html.parser -> html5lib

url = "https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=cs205"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html5lib') # html.parser / html5lib / lxml
print(soup)
```

```python
# 순위, 상호명, 업종, 주소, 사진 url 내용 출력
# select() vs select_one()
# select() : 매칭이 되는 여러 값들을 리스트에 담음
# select_one() : 매칭 되는 값 하나만 가져옴 

# select -> 하나의 열에 여러 선택자가 필요한 경우 -> or 대신 ',' 사용

# 속성값을 구해야 할 때 : [ ] 사용

ranking = soup.select_one('div.rankingList')
# print(ranking)
rankNum = ranking.select('ul > li > div > p.numTop, p.rankNum')
listName = ranking.select('ul > li > p.listName')
listType = ranking.select('ul > li > p.listType')
listArea = ranking.select('ul > li > p.listArea')
listPhoto = ranking.select('ul > li > p.listPhoto')

for t in range(0, len(rankNum)):
  for l in rankNum, listName, listType, listArea, listPhoto:
    print(l[t].get_text())
```

```python
# body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul
s = 'body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul'

ulTag = soup.select_one(s)
# print(ulTag)

for li in ulTag.select('li'):
  print('순위 : ', li.select_one('.numTop, .rankNum').string)
  print('상호명 : ', li.select_one('.restName').string)
  print('업종 : ', li.select_one('.listType').string)
  print('주소 : ', li.select_one('.listArea').string)
  # print('사진url : ', li.select_one('.restName a')['href'])
  print('사진url : ', li.select_one('.listPhoto a')['href'])
  # img의 경로지정 속성 : src (경로에 있는 이미지를 불러옴)
  # a태그의 url경로지정 속성 : href (해당 경로로 이동)
  print('=' * 20)
```

```python
# 데이터들을 list로 나타내는 작업
import numpy as np
# [{'순위' : '1', '상호' : 'XXX', '업종' : 'XXX', ...},
#   {'순위' : '2', '상호' : 'XXX', '업종' : 'XXX', ...},
#   {'순위' : '3', '상호' : 'XXX', '업종' : 'XXX', ...},
# ]


rank=[]
name=[]
sType=[]
addr=[]
photo=[]
for li in ulTag.select('li'):
  rank.append(li.select_one('.numTop, .rankNum').string + '위')
  name.append(li.select_one('.restName').string)
  sType.append(li.select_one('.listType').string)
  addr.append(li.select_one('.listArea').string)
  
# url에 대한 내용은 따로 별개의 list에 담기
  photo.append(li.select_one('.listPhoto a')['href'])

############################################################################
menuList = []
urlList = []

s = 'body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul'

ulTag = soup.select_one(s)
for li in ulTag.select('li'):
  menuList.append(  { '순위' : li.select_one('.numTop, .rankNum').string,
                      '상호명' : li.select_one('.restName').string,
                      '업종' : li.select_one('.listType').string,
                      '주소' : li.select_one('.listArea').string  } )
  
  urlList.append(li.select_one('.listPhoto a')['href'])
  # print('사진url : ', li.select_one('.restName a')['href'])
  # print('사진url : ', li.select_one('.listPhoto a')['href'])
# menuList
```

```python
# menuList를 가지고 순위를 index로 하는 DataFrame 생성
data = {'상호' : name,
        '업종' : sType,
        '주소' : addr}
df = pd.DataFrame(data, index=rank)
#########################################
menuDF = pd.DataFrame(menuList)
menuDF.set_index('순위', inplace=True)
menuDF
```

```python
# https://www.menupan.com/ + 담아둔 주소값
# 전화번호 + 상세주소 (구 주소) 가져오기
# 만들어둔 DF에 붙이기 + 기존 주소열 삭제
restAdd = []
restTel = []
url2 = 'https://www.menupan.com/'

for url in urlList:
  response = requests.get(url2 + url)
  soup = BeautifulSoup(response.content, 'html5lib') # html.parser / html5lib / lxml
  restTel.append(soup.select_one('.tel1').string)
  restAdd.append(soup.select_one('.add1').string)
  
menuDF['전화번호'] = restTel
menuDF['상세주소'] = restAdd
# menuDF.drop(columns='주소', inplace=True)
menuDF
```

```python
# 업종이 한식인 가게의 정보 출력
menuDF[menuDF['업종'] == '한식']
```

```python
# 특정 구에 있는 가게의 정보만 출력
# Pandas에서 Series의 특정 문자열 찾기 : str.contains()

menuDF[menuDF['상세주소'].str.contains('정자동')]
```
