# 1. 시리즈 (Series)

Series : 객체를 담을 수 있는 1차원 배열같은 자료구조

어떤 Numpy 자료형도 가능함



```python
# 시리즈 생성 방법
# 1-1. Dictionary (dict)
#   시리즈는 Dict 구조와 비슷함
dic = {'a' : 1, 'b' : 2, 'c' : 3, 'd' : 4}
series_dic = pd.Series(dic)

print(type(series_dic))
series_dic
print(series_dic)

# Dict의 Key = Index, Value = 값
```

```python
# Index와 값 확인
print(series_dic.index)
print(series_dic.values)
```

```python
# 1-2. List를 이용해 생성

obj = pd.Series([4, 5, -4, -5])
obj

print(obj.index)  # RangeIndex(start=0, stop=4, step=1) = range(4)
print(obj.values)

# RangeIndex(start=0, stop=4, step=1)
# Sample Data Frame을 만들 때 따로 Index 설정을 하지 않으면 0부터 시작하는 Index가 자동으로 부여 (start=0)
# 총 4개의 Index(stop = 4) 
# Index값이 1씩 늘어나도록 설정(step=1)
```

```python
obj2 = pd.Series([4, 5, -4, -5], index=['a', 'b', 'c', 'd'])
obj2
```

```python
# 1-3. Tuple을 이용해 생성
tup_data = ('Heim', '2022-12-07', '남', True)
series_data = pd.Series(tup_data, index = ['name', 'birthday', 'gender', 'T/F'])
print(series_data)
```

```python
# 1-4. 요소 선택 (Indexing)
print(series_data[0])
print(series_data['name'])
```

```python
# 여러 값을 Indexing
# 짝을 이루는 index와 값을 함께 반환

# 주의점
# 각각의 위치 or 이름으로 Indexing할 때, 대괄호 2개 사용해야함
print(series_data[[1, 2]])
print()
print(series_data[['birthday', 'gender']]) 
```

```python
# ':'를 이용해 범위를 Indexing 할 때는 대괄호 하나 사용
print(series_data[0:3])
print()
print(series_data['name':'gender'])
# Index의 이름으로 사용 할 때, 마지막 범위의 설정값 까지 포함 
# [1:3] : 2번째 Index까지
# ['name' : 'gender'] : 'gender'값 까지
```

```python
# Boolean 배열을 사용해 값 걸러내기 or 산술 연산 or 수학 함수 적용 가능
obj2 = pd.Series([4, 5, -4, -5], index=['a', 'b', 'c', 'd'])
print(obj2[obj2 > 0])
print()
print(obj2 * 2)
print()
print('b' in obj2)
print()
print('e' in obj2)
```

```python
ticket = {'busan' : 30000, 'jeju' : 50000, 'jeonju' : 40000, 'daejeon' : 55000}
ticketS = pd.Series(ticket)
ticketS
```

```python
city = ['seoul', 'busan', 'jeju', 'jeonju']
a = pd.Series(ticket, index=city)
a
```

```python
# 누락된 데이터 (NaN) 찾기
pd.isnull(a)
```

```python
a.isnull()
```

```python
# 정상 데이터 찾기 (누락되지 않은 데이터)
pd.notnull(a)
```

# DataFrame의 구조

Row (행) , Column(열)에 대한 Index를 가짐

```python
# Data Frame 생성 방법
# 1-1. Dictionary (Dict)
dict_data = {'c0': [1, 2, 3], 'c1': [4, 5, 6], 'c2' : [7, 8, 9], 'c3' : [10, 11, 12]}
df = pd.DataFrame(dict_data)

print(type(df))
print(df)

# 각각의 Column이 Series, Dict의 key값이 column의 이름이 됨
```

```python
# 1-2. List (1)
#   같은 길이의 List에 담긴 Dict / Numpy를 이용

data = {'time' : ['morning', 'afternoon', 'night'],
        'startHour' : [6, 12, 18],
        'behavior' : ['getup', 'lunch', 'gohome']
        }
data1 = pd.DataFrame(data)
data1
```

```python
# Column 순서 지정
pd.DataFrame(data, columns=['time', 'behavior', 'startHour'])
```

```python
# 1-3. List(2)
df = pd.DataFrame([[21, '남', '00대'], [22, '여', 'xx대']],
                  index = ['김김김', '박박박'],
                  columns = ['age', 'gender', 'school'])
print(df)
print()
print(df.index)
print()
print(df.columns)
```

```python
# 2-3. 행, 열 이름 변경 : 기본 방식
df.index = ['student1', 'student2']
df.columns = ['나이', '성별', '학교']
df

# Index와 column의 수에 맞추어 지정해주기
```

```python
# 2-4. 행, 열 이름 변경 : rename() 함수
# df명.rename(column = {변경 전 열 이름 : 변경 후 열 이름, ...}, inplace = True)
# inplace = True   => DataFrame에 바로 적용이 됨
df = pd.DataFrame([[21, '남', '00대'], [22, '여', 'xx대']],
                  index = ['김김김', '박박박'],
                  columns = ['age', 'gender', 'school'])
df
```

```python
# 열 이름 변경
df.rename(columns = {'age' : '연령', 'gender' : '성별', 'school' : '소속'}, inplace = True)
df
```

```python
# 행 이름 변경 => [index = ]
df.rename(index = {'김김김' : '김씨' , '박박박' : '박씨'}, inplace = True )
df
```

# 행 삭제 : drop() 함수

```python
exam_data = {'math' : [90, 80, 70], 'korean' : [98, 89, 95],
             'english' : [85, 95, 100], 'history' : [100, 90, 90]}

df = pd.DataFrame(exam_data, index = ['student1', 'student2', 'student3'])
df
```

```python
# drop('행 or 열의 이름' , axis = 0 or 1, inplace = True)
# 행 삭제시 axis=0 (기본값)
# 열 삭제시 axis=1 (필수입력)

df2 = df.copy()
# df2

df2.drop('student2', inplace=True)
df2
```

```python
# 열 삭제
# 2개 이상의 열 drop
#   => 대괄호 [] 안에 명시

df3 = df.copy()

df3.drop(['math', 'history'], axis=1, inplace=True)
df3
```

# Data Frame Indexing : loc() / iloc()

loc와 iloc는 가장 많이 사용되는 함수

  => 데이터 전처리 시 Indexing이 가장 많이 사용됨

loc는 이름(column명)으로, iloc는 숫자로 indexing

```python
city = pd.DataFrame({'Korea' : ['Seoul', 'Jeju', 'Bucheon'],
                     'US' : ['LA', 'Califonia', 'Boston'],
                     'China' : ['Shanhai', 'Beijing', 'Chingtao']},
                     index=['A', 'B', 'C']
                     )
city
```

```python
first = city.loc['A']
print(first)
print()
second = city.iloc[1]
print(second)
```

```python
city.ndim # 차원 (랭크)
```

```python
# 범위 Slicing
exam_data = {'name' : ['Oh', 'Woo', 'Ah', 'Ha', 'Kim', 'Lee', 'Park'],
             'math' : [90, 80, 70, 60, 50, 40, 30],
             'eng' : [98, 89, 95, 85, 45, 76, 77],
             'kor' : [94, 55, 66, 76, 56, 45, 34],
             'pt' : [100, 90, 99, 69, 79, 77, 86]
             }
df = pd.DataFrame(exam_data)
df
```

```python
# 요소 선택 
# set_index() : 특정한 열을 index로 부여함
exam_data = {'name' : ['Oh', 'Woo', 'Ah', 'Ha', 'Kim', 'Lee', 'Park'],
             'math' : [90, 80, 70, 60, 50, 40, 30],
             'eng' : [98, 89, 95, 85, 45, 76, 77],
             'kor' : [94, 55, 66, 76, 56, 45, 34],
             'pt' : [100, 90, 99, 69, 79, 77, 86]
             }
df = pd.DataFrame(exam_data)
df.set_index('name', inplace=True)
df
```

```python
# loc, iloc 행/열 선택과정에서
# 두 가지 이상의 행 / 열을 선택시 [[]] 사용
# []만 사용하면 그 좌표의 값(요소)이 나옴 
df.loc['Ha', 'kor']
```

```python
# 해당하는 요소 하나 반환
df.iloc[3, 2]
```

```python
# Ah의 eng, kor 점수를 4가지 방법을 이용해 출력
# 1
a = df.loc['Ah', ['eng', 'kor']]
print(a)
print()
# 2
a2 = df.iloc[2, [1, 2]]
print(a2)
print()
# 3
a3 = df.loc['Ah', 'eng':'kor']
print(a3)
print()
#4
a4 = df.iloc[2, 1:3]
print(a4)
```

```python
# Oh, Woo의 kor, pt점수를 4가지 방법을 이용해 출력
ow1 = df.loc[['Oh', 'Woo'], ['kor', 'pt']]
print(ow1)
print()
ow2 = df.loc['Oh':'Woo', 'kor' : 'pt']
print(ow2)
print()
ow3 = df.iloc[[0,1], [2, 3]]
print(ow3)
print()
ow4 = df.iloc[0:2, 2:4]
print(ow4)
```

```python
# 행, 열 추가
# 열 추가 : df명['새로운 열이름'] = 값 의 형태
#   새로운 열 이름이 아닌 기존의 열 이름을 입력하면 기존 값들은 다 덮어씌워짐
exam_data = {'name' : ['Oh', 'Woo', 'Ah', 'Ha', 'Kim', 'Lee', 'Park'],
             'math' : [90, 80, 70, 60, 50, 40, 30],
             'eng' : [98, 89, 95, 85, 45, 76, 77],
             'kor' : [94, 55, 66, 76, 56, 45, 34],
             'pt' : [100, 90, 99, 69, 79, 77, 86]
             }
df = pd.DataFrame(exam_data)
df['music'] = 80    # 모든 행에 대해 같은 값이 입력
df
```

```python
exam_data = {'name' : ['Oh', 'Woo', 'Ah', 'Ha', 'Kim', 'Lee', 'Park'],
             'math' : [90, 80, 70, 60, 50, 40, 30],
             'eng' : [98, 89, 95, 85, 45, 76, 77],
             'kor' : [94, 55, 66, 76, 56, 45, 34],
             'pt' : [100, 90, 99, 69, 79, 77, 86]
             }
df = pd.DataFrame(exam_data)

# 요소 값 변경
df.loc[3] = 10                                # 단일값 입력시 행 전체가 같은 값으로 채워짐
df.loc[4] = ['Wa', 100, 100, 100, 100]        # 기존의 행 / 열의 수에 맞게 요소 변경 / 추가
df.loc[7] = ['asdf', 50, 50, 50, 50]          # 새로운 행 추가
df
```

```python
# 행 / 열 전환 : 전치행렬 transpose(), .T
exam_data = {'name' : ['Oh', 'Woo', 'Ah', 'Ha', 'Kim', 'Lee', 'Park'],
             'math' : [90, 80, 70, 60, 50, 40, 30],
             'eng' : [98, 89, 95, 85, 45, 76, 77],
             'kor' : [94, 55, 66, 76, 56, 45, 34],
             'pt' : [100, 90, 99, 69, 79, 77, 86]
             }
df = pd.DataFrame(exam_data)
print()

df = df.transpose()
print(df)
print()

df = df.T
print(df)

# 전치 두번 -> 원래 상태로 돌아옴
```
