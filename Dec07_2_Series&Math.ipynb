# 1-2. 시리즈 vs 시리즈

시리즈와 시리즈 간의 연산은 같은 인덱스를 가진 요소끼리 연산하여 새로운 시리즈 반환

같은 Index를 가지지만 순서가 다른 두 시리즈를 생성

```python
student1 = pd.Series({'국어' : 100, '영어' : 80, '수학' : 90})
student2 = pd.Series({'수학' : 90, '영어' : 70,  '국어' : 80})

print(student1)
print(student2)
```

```python
# 사칙연산
sum = student1 + student2
print(sum)
mins = student1 - student2
print(mins)
mult = student1 * student2
print(mult)
div = student1 / student2
print(div)
```

```python
# 그 결과를 데이터프레임으로 합치기 (시리즈 -> 데이터프레임)
# Series를 데이터 프레임으로 변환 시, Series는 [ ]안에 합치면, 각 Index는 열의 이름이 됨
result = pd.DataFrame([sum, mins, mult, div], 
                      index = ['덧셈', '뺄셈', '곱셈', '나눗셈'])
result
```

# 1-3. NaN 값이 있는 Series 연산

```python
student1 = pd.Series({'국어' : np.nan, '영어' : 80, '수학' : 90})
student2 = pd.Series({'수학' : 90, '국어' : 70, })
print(student1)
print()
print(student2)
print()

# 이 둘을 더한 결과 예상
# 국어 70 영어 80 수학 180?
print(student1 + student2)
# -> NaN과 연산 / 없는 Index와의 연산은 NaN으로 반환이 됨
```

# 1-4. 연산 함수 (add(), sub(), mul(), div()) 와 fill_value()

연산 부호로 연산이 가능하지만, 연산 함수도 존재함

연산결과가 NaN이 되는 두가지 경우는 fill_value() 옵션으로 값을 대체 가능

```python
add_data = student1.add(student2, fill_value=0)
sub_data = student1.sub(student2, fill_value=0)
mul_data = student1.mul(student2, fill_value=0)
div_data = student1.div(student2, fill_value=0)

result = pd.DataFrame([add_data, sub_data, mul_data, div_data],
                      index = ['덧셈', '뺄셈', '곱셈', '나눗셈'])
result

# inf = infinite
# '영어' Index를 연산할 때 NaN을 0으로 대체해서 연산
#   => 80 / 0 = inf
```
