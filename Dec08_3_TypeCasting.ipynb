# 1. 데이터 표준화

# 1-1. 단위 환산 : round() / 반올림


```python
t = sns.load_dataset('titanic')
t.head()
```

```python
# round() : 반올림 함수
# fare 값을 소수점 2자리 까지 표기
t['fare'] = t['fare'].round(2) # () 안의 숫자 = 표현할 소수점의 자리수
t.head()
```

```python
# round()의 사용 : df명.round(소수점자리수) / round(df명, 소수점자리수)
# round()의 default값 : 0 (첫째 자리에서 반올림 / 소수점 아래 수 표기 x)
print(round(1.6667))
print(round(1.6667, 3))
```

```python
# .xls or .xlsx 확장자 파일 불러오기 
# read_excel()
a = pd.read_excel('myStudent.xlsx')
a.head()

# 불러오기가 되지 않을 때
# !pip install --upgrade xlrd
```

```python
# 국어, 영어, 수학점수를 다 더한 값의 평균값 계산
# df결과 옆에 추가하기
a['avg'] = ((a['kor'] + a['eng'] + a['math']) / 3).round(2)
a.head()
```

```python
a.dtypes
```

```python
a['gender'] = a['gender'].astype('category')
a.dtypes
```

```python
a[['group', 'education']] = a[['group', 'education']].astype('category')
a.dtypes
```

```python
a[['gender', 'group', 'education']].describe()
```

```python
# replace() : 값 변경
# 다양한 값을 한번에 변경 가능
# replace( {변경전 값 : 변경후 값, ...} )
# a['gender']
a['gender'].replace({'f' : 'female', 'm' : 'male'}, inplace=True)
a['gender'].unique()
```

# 2. 범주형(카테고리) 데이터 처리

# 구간 분할 : np.histogram(), pd.cut()



```python
aa = pd.DataFrame(a)
aa.head()
```

```python
# np.histogram() : np.histogram(나눌 리스트, bins=나눌 구간)
# 2개의 값 도출
#   count : 각 구간에 속하는 값의 개수
#   bins_dividers : 경계값의 내용을 리스트에 담아 반환
count, bins_dividers = np.histogram(aa['avg'], bins=3)
print(count, '\n', bins_dividers)
```

```python
# pd.cut() : 자동으로 범주형 변수를 생성해줌, 간단함
bin_names = ['low', 'middle', 'high']
aa['aa_bins'] = pd.cut(x = aa['avg'],             # 데이터의 배열
                       bins = bins_dividers,      # 경계값 리스트
                       labels=bin_names,          # 구간명
                       include_lowest=True)       # 첫 경계값 포함 여부 (구간의 하위값)

aa.head()
```
